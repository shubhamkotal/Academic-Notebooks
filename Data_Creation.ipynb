```python
def correct(path):
    jan=pd.read_csv(path)
    try:
        jan.columns =['Ticker', 'Date', 'Time', 'Open','High','Low','Close'] #,'dlt1','dlt2']
    except:
        pass
    try:
        jan.columns =['Ticker', 'Date', 'Time', 'Open','High','Low','Close','dlt1'] #'dlt2']
        jan=jan.drop('dlt1',axis=1)
    except:
        pass
    try:
        jan.columns =['Ticker', 'Date', 'Time', 'Open','High','Low','Close','dlt1','dlt2']
        jan=jan.drop(['dlt1','dlt2'],axis=1)
    except:
        pass  
    jan=jan[jan['Time'].str.contains(timings)]
    # jan=jan[jan['Time'] != '09:09']
    return jan
```
