```python
# Divide in to train and test
train_target=train["amount_spent_per_room_night_scaled"]
train1=train.drop('amount_spent_per_room_night_scaled', axis=1)
#from sklearn.model_selection import train_test_split  
train2,val2,train_target2,val_target2 = train_test_split(train1, train_target, test_size=0.20)  
```

```python
# Import the model we are using
from sklearn.ensemble import RandomForestRegressor
# Instantiate model with 250 decision trees
rf = RandomForestRegressor(n_estimators = 100,max_features=4, random_state = 42)
# Train the model on training data
rf.fit(train2,train_target2)
```

```python
from math import sqrt
```

```python
from sklearn.metrics import mean_squared_error

predictions1 = rf.predict(val2)

meanSquaredError=mean_squared_error(val_target2, predictions1)
print("MSE:", meanSquaredError)
rootMeanSquaredError = sqrt(meanSquaredError)
print("RMSE:", rootMeanSquaredError)
```

```python
#Saving predictions

test_pred2= rf.predict(test)
test_pred2
```

```python
q75, q25 = np.percentile(testing['amount_spent_per_room_night_scaled'], [75 ,25])
iqr = q75-q25
print("IQR",iqr)
upper = q75 + (1.5*iqr)
print("Upper",upper)
lower = q75 - (1.5*iqr)
print("Lower",lower)
```

```python
#replace outliers with upper wishker
testing['amount_spent_per_room_night_scaled'] = testing['amount_spent_per_room_night_scaled'].clip(upper=upper)
```
