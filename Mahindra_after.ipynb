```python
#split data
from sklearn.model_selection import train_test_split
```

```python
# Divide in to train and test
train_target=train["amount_spent_per_room_night_scaled"]
train1=train.drop('amount_spent_per_room_night_scaled', axis=1)
#from sklearn.model_selection import train_test_split  
train2,val2,train_target2,val_target2 = train_test_split(train1, train_target, test_size=0.20)  
```

```python
from sklearn.preprocessing import StandardScaler
num_values=train2.select_dtypes(['int64','float64','int32','float32']).columns
scaler = StandardScaler()
scaler.fit(train2[num_values])
#scaler.fit(val2[num_values])
```

```python
# Import the model we are using
from sklearn.ensemble import RandomForestRegressor
# Instantiate model with 250 decision trees
rf = RandomForestRegressor(n_estimators = 1000,max_features=4, random_state = 42,max_depth=2)
# Train the model on training data
rf.fit(train2,train_target2)
```

```python
from math import sqrt
```

```python
from sklearn.metrics import mean_squared_error

predictions11 = rf.predict(train2)

meanSquaredError=mean_squared_error(train_target2, predictions11)
print("MSE:", meanSquaredError)
rootMeanSquaredError = sqrt(meanSquaredError)
print("RMSE:", rootMeanSquaredError)
```

```python
from sklearn.metrics import mean_squared_error

predictions1 = rf.predict(val2)

meanSquaredError=mean_squared_error(val_target2, predictions1)
print("MSE:", meanSquaredError)
rootMeanSquaredError = sqrt(meanSquaredError)
print("RMSE:", rootMeanSquaredError)
```

```python
feat_importances = pd.Series(rf.feature_importances_, index=train2.columns)
feat_importances = feat_importances.sort_values(ascending=False)
feat_importances.nlargest(20).plot(kind='barh')
```
